```python
def distance(A,B):
    dis=np.sqrt(np.sum(np.square((A-B))))
    return dis
```

```python
def totaldistance(coordinates):
    dist=0
    for i in range(130):
        dist=dist+distance(coordinates[i,:],coordinates[i+1,:])
    dist=dist+distance(coordinates[0,:],coordinates[-1,:])
    return dist
```
