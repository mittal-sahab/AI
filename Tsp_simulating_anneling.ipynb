```python
import numpy
import matplotlib.pyplot as plt

x=np.array(np.random.rand(20)*100)

y=np.array(np.random.rand(20)*100)
fig=plt.scatter(x,y)
coordinates=np.array([x,y]).T
print(coordinates)
print()
coordinates[[1,2]]=coordinates[[2,1]]
print(coordinates)
```

```python
def distance(A,B):
    dis=np.sqrt(np.sum(np.square((A-B))))
    return dis
```

```python
def totaldistance(coordinates):
    dist=0
    for i in range(19):
        dist=dist+distance(coordinates[i,:],coordinates[i+1,:])
    dist=dist+distance(coordinates[0,:],coordinates[-1,:])
    return dist
```

```python
cost0= totaldistance(coordinates)

T=30
factor=0.99
T_init=T
for i in range(600):
    print(i, 'cost=',cost0)
    
    T=T*factor
    for j in range(500):
        r1, r2 = numpy.random.randint(0,20, size=2)
        coordinates[[r1,r2]]=coordinates[[r2,r1]]
        cost1 = totaldistance(coordinates)
        if(cost1<cost0):
            cost0=cost1
        else:
            x = np.random.uniform()
            if x < np.exp((cost0-cost1)/T):
                cost0 = cost1
            else:
                coordinates[[r1,r2]]=coordinates[[r2,r1]]

                
                
```
